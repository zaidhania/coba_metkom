```python
#SIMO
import numpy as np #mengimport modul
import pylab as plt #mengimport modul

#Time interval
dt = 0.01 #mengatur resolusi grafik agar halus
t = np.arange(-5,5+dt,dt) #interval waktu -5<t<5

#input signal
x1 = t #input signal x(t)

#Output signal 
y1 = np.sin(x1) #output signal y1(t)
y2 = x1 #output signal y2(t)

#Plot input signal
fig,ax = plt.subplots() #plot untuk memisahkan grafik
plt.title("Input signal") #memberikan judul grafik
plt.plot(t,x1, '-k') #memplot grafik
plt.grid() #memberikan grid pada grafik

#Plot output signal
fig,ax = plt.subplots() #plot untuk memisahkan grafik agar tidak terpengaruh grafik lainnya
plt.plot(t,y1,'-k',label="$y_1(t)$") #memplot signal y1(t)
plt.plot(t,y2,'--',label="$y_2(t)$") #memplot signal y2(t)
plt.grid() #memberikan grid pada grafik
plt.legend() #memberikan label pada grafik
plt.title("Output signals") #memberikan judul grafik
```

```python
#MIMO
import numpy as np
import pylab as plt

#Time interval
dt = 0.01 #mengatur resolusi grafik agar halus
t = np.arange(-5,5+dt,dt) #interval waktu -5<t<5

#Input signal x1 and x2
a = t
x1 = np.sin(a)
x2 = a

#Output signal y1 dan y2
y1 = x1*x2  # Output signal y1(t)
y2 = -x1*x2 # Output signal y2(t)

#Plot Input signal
fig,ax = plt.subplots()# plot pemisah
plt.plot(t,x1,'-k',label="$x_1(t)$")# plotting input signal
plt.plot(t,x2,'--',label="$x_2(t)$")# plotting input signal
plt.legend(loc="upper center")# memberikan label pada grafik plot
plt.title("Input signal")  # memberikan judul pada grafik plot
plt.grid() #memberikan grid pada plot

#Plot output signal
fig,ax = plt.subplots()# plot pemisah
plt.plot(t,y1,'-k',label="$y_1(t)$")# plotting output signal y1(t)
plt.plot(t,y2,'--',label="$y_2(t)$")# plotting output signal y2(t)
plt.legend(loc="upper center")# memberikan label pada grafik plot
plt.title("Output signal")# memberikan judul pada grafik plot
plt.grid() #memberikan grid pada plot
plt.show()#menampilkan grafik plot
```

```python
#MISO
import numpy as np  #mengimport modul 
import pylab as plt #mengimport modul 

#Time interval
dt = 0.01 #waktu interval digunakan untuk mengatur resolusi grafik
t = np.arange(-5,5+dt,dt) #penambahan waktu interval

#Input signal 
a=t 
x1 = np.sin(a)*a #Input signal x1(t)
x2 = -np.sin(a)*(a) #Input signal x2(t)

#Mendefinisikan interval
t1 = np.arange(-5,-0.01+dt,dt)   #Mendefinisikan interval waktu pertama    
t2 = [0.]  #Mendefinisikan interval waktu kedua
t3 = np.arange(0.01,5+dt,dt) #Mendefinisikan interval waktu pertama
T  = np.concatenate((t1,t2,t3)) #menggabungkan dari t1,t2,t3 menjadi 1 bagian

# Proses mencari output yang bersyarat
y_1a = -np.sin(t1)*(t1)  #menggunakan t1 karena output < 0 termasuk dalam interval t1
y_1b = np.zeros_like(t2) #menggunakan zeros karena nilai akan selalu 0
y_1c = np.sin(t3)*t3     #menggunakan t3 karena output akan >= 0 termasuk dalam interval t3
y = np.concatenate((y_1a,y_1b,y_1c))#menggabungkan dari y_1a,y_1b,y_1c menjadi 1 bagian


#Plot input signals
fig,ax = plt.subplots()#plot pemisah
plt.plot(t,x1,'-k',label='$x_1(t)$')#plotting sinyal masukan x1
plt.plot(t,x2,'--',label='$x_2(t)$')#plotting sinyal masukan x2
plt.legend(loc="upper center")#memberikan label pada grafik plot
plt.title("Input signal")  #memberikan judul pada grafik plot
plt.grid() #memberikan grid pada plot

# Plot output signals
fig,ax = plt.subplots()# plot pemisah
plt.plot(T,y,'-k',label='$y(t)$')# plotting sinyal keluaran 
plt.legend(loc="upper center")# memberikan label pada grafik plot
plt.title('Output signal')# memberikan judul pada grafik 
plt.grid() #memberikan grid pada grafik
plt.show()#menampilkan grafik
```

```python
import numpy as np  #mengimport modul
import pylab as plt #mengimport modul

#Time Interval
dt = 0.01 #waktu interval digunakan, semakin kecil grafik semakin halus
t  = np.arange(-5,5+dt,dt) #penambahan tiap waktu dari interval yang ada yaitu 0.01 tiap data

#Mendefinisikan fungsi heaviside
H1=np.heaviside(t,1)  #mendefinisikan fungsi heaviside
H2=np.heaviside(t-5,1)#mendefinisikan fungsi heaviside
h=np.sin(3*np.pi*t)*(H1-H2) #rumus impulse respon 
x=np.cos(2*np.pi*t)*(H1-H2) #input signal


#Plot input signal
fig,ax = plt.subplots() #plot pemisah
plt.plot(t,x,label='$x(t)$') #memplott input signal
plt.legend(loc="upper left") #memberikan label pada grafik plot
plt.title("Input signal") #memberikan judul pada grafik plot
plt.grid() #memberikan grid pada grafik

#Plot output signal
fig,ax = plt.subplots() #plot pemisah
plt.plot(t,h,label='$h(t)$') # plotting impulse respone sebagai output signal
plt.legend(loc="upper left") # memberikan label pada grafik plot
plt.title("Output signal") # memberikan judul pada grafik plot
plt.grid() #memberikan grid pada grafik
plt.show() #memunculkan grafik 
```

```python
import numpy as np  # import modul
import pylab as plt # import modul

#Time interval
dt = 0.01 # waktu interval digunakan untuk mengatur resolusi grafik
t  = np.arange(-5,5+dt,dt)#penambahan tiap waktu dari interval yang ada yaitu 0.01 tiap data

# Definisi fungsi heaviside
H  = np.heaviside(t,1) 
H1 = np.heaviside(t+2,1)
H2 = np.heaviside(t-3,1)

h  = 5*np.exp(-t)*H #impulse response

x = np.cos((np.pi)*t)*(H1-H2) #Input signal
y = np.convolve(x,h,'same')*dt #hasil konvolusi sinyal masukan dan sistem
                            # 'same' untuk mengembalikan data yang panjangnya sama dengan vektor terpendek 

ty= np.arange(0,10+dt,dt) # rentang waktu dari grafik hasil konvolusi 0 hingga 10 karena sifat konvolusi yang membalik data

#memplot signal input
fig,ax = plt.subplots() #plot pemisah
plt.plot(t,x,label='$x(t)$')#memplotting input signal
plt.legend(loc="upper left")#memberikan label pada grafik
plt.title("input signal")# memberikan judul pada grafik
plt.grid()# memberikan grid pada grafik

#memplot signal output 
fig,ax = plt.subplots() #plot pemisah
plt.plot(t,h,label='$h(t)$') #memplotting impulse respone sebagai output signal
plt.legend(loc="upper left") #memberikan label pada grafik
plt.title("Output signal") #memberikan judul pada grafik
plt.grid() #memberikan grid pada grafik


#memplot grafik convolution
fig,ax = plt.subplots() #plot pemisah
plt.plot(ty,y) #memplotting convolution signal
plt.title("convolution signal") #memberikan judul pada grafik
plt.grid() #memberikan grid pada grafik 
plt.show() #memunculkan grafik 
```

```python
import numpy as np  #mengimport modul 
import pylab as plt #mengimport modul 

#time interval
dt = 0.01 #waktu interval digunakan untuk mengatur resolusi grafik
#interval waktu pertama
t1 = np.arange(0,1+dt,dt)#penambahan tiap waktu dari interval yang ada yaitu 0.01 tiap data
#interval waktu kedua
t2= np.arange(1.01,2+dt,dt)#penambahan tiap waktu dari interval yang ada yaitu 0.01 tiap data
t = np.concatenate((t1,t2))# menggabungkan dari t1,t2 menjadi 1 bagian

x1=t1 # definisi persamaan sinyal untuk interval t1
x2=2-t2 # definisi persamaan sinyal untuk interval t2
x = np.concatenate((x1,x2))# menggabungkan dari h1,h2 menjadi 1 bagian

# Plot input signals
fig,ax = plt.subplots()# plot pemisah
plt.plot(t,x,label='$x(t)$')#memplotting sinyal masukan x1
plt.legend(loc="upper right")#memberikan label pada grafik plot
plt.title("Input signal")  #memberikan judul pada grafik plot
plt.grid() #memberikan grid pada plot

#plot output signal
to=np.arange(0,1+dt,dt)#penambahan tiap waktu dari interval yang ada yaitu 0.01 tiap data
h=1-to # definisi persamaan sinyal untuk interval to
fig,ax = plt.subplots()#memplot pemisah
plt.plot(to,h,label='$h(t)$')# plotting sinyal impuls sebagai output
plt.legend(loc="upper right")#memberikan label pada grafik plot
plt.title("Output signal")  #memberikan judul pada grafik plot
plt.grid() #memberikan grid pada plot

y=np.convolve (x,h)*dt #hasil konvolusi sinyal masukan dan sistem
ty=np.arange (0,3+dt,dt)#rentang waktu dari grafik hasll konvolusi dari 0 hingga 3

#plot konvolusi
fig,ax = plt.subplots() #plot pemisah
plt.plot(ty,y)# plotting convolution signal
plt.title('convolution signal')# memberikan judul pada grafik plot
plt.grid() #memberikan grid pada plot
plt.show()# menampilkan plot
```
